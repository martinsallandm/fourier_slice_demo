# Fourier Slice theorem

O Fourier Slice Theorem diz que a transformada de Fourier das projeções 1D em um ângulo $\beta$ na transfomada Radon é igual aos valores da tranaformada 2D de Fourier ao long da linha com mesmo ângulo $\beta$



## Transformada Radon e projeções

- ProjeçÕes são integrais (informalmente uma soma ou um acúmulo) dos valores de uma função 2D (imagem) ao long de um ânglulo $\beta$
- Cada ângulo provê várias projeções que são <b>paralelas</b> no mesmo ângulo $\beta$

<img src="proj.png"  width="400">

- Na figura a seguir as linhas paralelas representam as projeções e a curva azul são os valores da integral da imagem ao long de cada projeção
- Se gerarmos estas curvas para vários ângulos e as enfileirarmos como colunas em umas outra imagem, esta imagem resultante se chama "sinograma". Um sinograma é o que chamamos de transformada Radon de uma imagem. Ver figura a seguir

<img src="sinogram.png"  width="300">

## Matemática envolvida

### Radon e projeções

- Formalmente, a transformada Radon de uma imagem $f(x,y)$ é dada por $R[f](\beta,t)$ e expressa como

$\int_{-\infty}^{\infty}{f( t\,sin(\beta) + \tau\,cos(\beta), t\,cos(\beta) - \tau\,sin(\beta) ) \, \mathrm{d}\tau}$

- Isto corresponde à integral de $f(x,y)$ ao long de uma linha paramétrica com ângulo $\beta$ passando pelo ponto $(t\,sin(\beta), - t\,cos(\beta))$. Parametricamente esta linha é dada por $(t\,sin(\beta) + \tau\,cos(\beta), t\,cos(\beta) - \tau\,sin(\beta))$)


<img src="geometry.png"  width="300">


- Deste modo, $R[f](\beta, \dot )$ é uma função 1D que pode ser representada no domínio da frequencia como qualquer outra função

### Teorema de Fourier Slice

#### Dominio espacial

- Vamos denominar $r_\beta(t)$ a função indicada a pouco, ou seja $r_\beta(t) = R[f](\beta,t)$.

- Pela definição da transformada Radon,


$r_\beta(t) = \int_{-\infty}^{\infty}{f(t\,sin(\beta) + \tau\,cos(\beta), t\,cos(\beta) - \tau\,sin(\beta)) \, \mathrm{d}\tau}$
    

- Podemos calcular a transformada de Fourier de $r_\beta(t)$ como

$R_\beta(\omega) = \int_{-\infty}^{\infty}{r_\beta(t)\,e^{-j\,\omega\,t}\mathrm{d}t}$ 

- Se substituirmos a definição de $r_\beta(t)$ obtemos

<font color='red'>

$R_\beta(\omega) = \int_{-\infty}^{\infty}{  \int_{-\infty}^{\infty}{f(  t\,sin(\beta) + \tau\,cos(\beta), t\,cos(\beta) - \tau\,sin(\beta) ) \, \mathrm{d}\tau}  \,e^{-j\,\omega\,t}\mathrm{d}t}$ 

</font>

#### Dominio da freqüência

- Agora podemos expressar a transformada de Fourier 2D de $f(x,y)$

$F(k_1, k_2) = \int_{-\infty}^{\infty}\int_{-\infty}^{\infty}{  f(x,y) \, e^{-j\,(k_1\,x + k_2 \, y)} \mathrm{d}x\mathrm{d}y}$ 


- De acordo com o terema de Fourier Slice, os valores desta tranformada quando ao long da linha de ângulo $\beta$ devem ser iguais a $R_\beta(\omega)$

- Parametrizando $k_1$ e $k_2$ na referida linha, podemos fazer (chamando o parâmetro covenientemente de $\omega$)

$k_1 = \omega \, sin(\beta)$

$k_2 = \omega \, cos(\beta)$

- A relação entre $x,y$ e $t, \tau$ são dadas pela projeção de ângulo $\beta$, ou seja (como definido na geometria anteriormente)

$x(t, \tau) = t\,sin(\beta) + \tau\,cos(\beta)$

$y(t, \tau) = t\,cos(\beta) - \tau\,sin(\beta)$


- agora podemos substituir $k_1$, $k_2$, $t$ e $\tau$ na tranformada 2D de Fourier e obter:


 $\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}{  f(t\,sin(\beta) + \tau\,cos(\beta),t\,cos(\beta) - \tau\,sin(\beta)) \, e^{-j\,(\omega\,sin(\beta)\,(t\,sin(\beta) + \tau\,cos(\beta)) + \omega\,cos(\beta) \, (t\,cos(\beta) - \tau\,sin(\beta)))} |J(x,y)|\mathrm{d}t\mathrm{d}\tau}$ 


- A jacobiana da tranformação é data por (cujo determinante é igual a 1)

$\left[  \begin{array}{cc}
cos(\beta) & sin(\beta) \\
-sin(\beta) & cos(\beta) \\
\end{array}
\right]$

- Simplificando a transformada 2D anterior obtemos

<font color='red'>
    
 $\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}{  f(t\,sin(\beta) + \tau\,cos(\beta),t\,cos(\beta) - \tau\,sin(\beta)) \, e^{-j\,\omega\,t } \mathrm{d}t\mathrm{d}\tau}$ 
 
</font>

- que é exatamente a tranformada de Fourier 1D da projeção em $\beta$


```python
def reshape(I):
    if I.shape[0]>I.shape[1]:
        dif = I.shape[0]-I.shape[1]
        I = I[dif//2:-dif//2,:]    
    else:
        dif = I.shape[1]-I.shape[0]
        I = I[:dif//2:-dif//2]            
    
    return I


def computeProjection(I, angle):
    
    Isq = reshape(I)
    
    imgSize = Isq.shape[0]
    diagSize = np.sqrt(2.0)*imgSize

    Irotated = imutils.rotate(Isq,angle=angle)
    
    return Irotated
    
def generate2DSineWave(imgSize, w0, angle):
    
    [x, y] = np.meshgrid(np.linspace(-np.pi, np.pi, imgSize), np.linspace(-np.pi, np.pi, imgSize))
    
    I = np.sin(0.2*np.sin(angle)*w0*x + 0.2*np.cos(angle)*w0*y)
    
    return I    


def generateFourierMask(imgSize, angle):

    [x, y] = np.meshgrid(np.linspace(-1.0, 1.0, imgSize), np.linspace(-1.0, 1.0, imgSize))

    I = ((x**2.0 + y**2.0)<1.0)*((-np.arctan2(y,x)<angle-np.pi)+(np.arctan2(y,-x)<angle-np.pi))
    
    return I
    
```

```python
def plotQuote(ax, p1, p2, text, quotaSize = 2.0):

    angle = np.arctan2(p2[1]-p1[1], p2[0]-p1[0])

    p1l = (p1[0] + quotaSize*np.cos(angle+np.pi/2.0), p1[1] + quotaSize*np.sin(angle+np.pi/2.0))
    p2l = (p2[0] + quotaSize*np.cos(angle+np.pi/2.0), p2[1] + quotaSize*np.sin(angle+np.pi/2.0))
    
    _cl = ((p1l[0]+p2l[0])/2.0, (p1l[1]+p2l[1])/2.0)
    #_cl = (_c[0] + quotaSize*np.cos(angle+np.pi/2.0), _c[1] - quotaSize*np.sin(angle+np.pi/2.0))
    
    ax.plot([(p1l[0]+p1[0])/2.0, (p2l[0]+p2[0])/2.0],[(p1l[1]+p1[1])/2.0, (p2l[1]+p2[1])/2.0],'-k',alpha=0.1)
    ax.plot([p1[0], p1l[0]],[p1[1], p1l[1]],'-k',alpha=0.1)
    ax.plot([p2[0], p2l[0]],[p2[1], p2l[1]],'-k',alpha=0.1)
    ax.text(_cl[0], _cl[1], text, ha='center', va='center', rotation=angle*180.0/np.pi)
    
    
def drawSineWare(ax, I, pos, angle):
    
    im = ax.imshow(I, interpolation='none',
                   origin='lower',
                   clip_on=True)

    trans_data = mtransforms.Affine2D().rotate_deg(180.0*angle/np.pi).translate(pos[0], pos[1]) + ax.transData
    im.set_transform(trans_data)    
    
    
    
    
    
    
def plotState(axs, I, IFrr, sino, angle, w0):
    
    # source and detector slask (distance to object)
    sourceSlack = 1.2
    detSlack = 1.2
    quotaSize = 12.0
    
    NRays = 10
    
    imgSize = I.shape[0]
    imgCenter = (imgSize/2.0, imgSize/2.0)
    
    diagSize = np.sqrt(2.0)*imgSize
    
    # plot image and source position
    axs[0].plot([0, imgSize, imgSize, 0, 0],[0, 0, imgSize, imgSize, 0],'b')
    
    p1 = (imgCenter[0] + sourceSlack*diagSize*np.cos(angle), imgCenter[1] +  sourceSlack*diagSize*np.sin(angle))
    p2 = (p1[0] + diagSize*np.cos(angle+np.pi/2.0)/2.0, p1[1] + diagSize*np.sin(angle+np.pi/2.0)/2.0)
    p3 = (p1[0] - diagSize*np.cos(angle+np.pi/2.0)/2.0, p1[1] - diagSize*np.sin(angle+np.pi/2.0)/2.0)
    axs[0].plot([0, imgSize, imgSize, 0, 0],[0, 0, imgSize, imgSize, 0],'b')
    axs[0].plot([p2[0], p3[0]],[p2[1], p3[1]],'k')
    
    plotQuote(axs[0], p2, p3, 'source', quotaSize=quotaSize)    
    
    plotQuote(axs[0], [0, 0], [imgSize, 0], 'imgSize', quotaSize=-quotaSize)
    
    
    # plot xRays
    for i in range(NRays+1):
        delta = diagSize*(-NRays/2.0 + i)/NRays
        
        p1 = (imgCenter[0] + sourceSlack*diagSize*np.cos(angle), imgCenter[1] + sourceSlack*diagSize*np.sin(angle))
        p2 = (p1[0] + delta*np.cos(angle+np.pi/2.0), p1[1] + delta*np.sin(angle+np.pi/2.0))
        p3 = (p2[0] - 2.0*detSlack*diagSize*np.cos(angle), p2[1] - 2.0*detSlack*diagSize*np.sin(angle))
        axs[0].plot([p2[0], p3[0]],[p2[1], p3[1]],'k',alpha=0.1)
    
    
    
    # plot detector
    _a = angle+np.pi
    p1 = (imgCenter[0] + detSlack*diagSize*np.cos(_a), imgCenter[1] + detSlack*diagSize*np.sin(_a))
    p2 = (p1[0] + diagSize*np.cos(_a+np.pi/2.0)/2.0, p1[1] + diagSize*np.sin(_a+np.pi/2.0)/2.0)
    p3 = (p1[0] - diagSize*np.cos(_a+np.pi/2.0)/2.0, p1[1] - diagSize*np.sin(_a+np.pi/2.0)/2.0)
    axs[0].plot([p2[0], p3[0]],[p2[1], p3[1]],'k')
    
    plotQuote(axs[0], p2,p3, 'detector', quotaSize=quotaSize)
    
    # draw image
    axs[0].imshow(I[::-1,::-1], cmap='gray')

    
    proj = np.sum(imutils.rotate(I,(180.0/np.pi)*angle),axis=0)
    
    _a = angle+np.pi
    p1 = (imgCenter[0] + detSlack*diagSize*np.cos(_a), imgCenter[1] + detSlack*diagSize*np.sin(_a))
    NP = len(proj)
    PROJsx = []
    PROJsy = []
    for i, p in enumerate(proj):
        x = imgSize*(-NP/2.0+i)/NP
        y = p/255.0
        
        xr = np.cos(-angle+np.pi/2.0)*x + np.sin(-angle+np.pi/2.0)*y
        yr = -np.sin(-angle+np.pi/2.0)*x + np.cos(-angle+np.pi/2.0)*y
        
        PROJsx.append( p1[0] + xr )
        PROJsy.append( p1[1] + yr )
        
    axs[0].plot(PROJsx, PROJsy)
       
    
    Isine = generate2DSineWave(N, w0*imgSize/2.0, 0.0)
    
    p1 = (imgCenter[0] + (detSlack*imgSize+detSlack*diagSize)*np.cos(_a), imgCenter[1] + (detSlack*imgSize+detSlack*diagSize)*np.sin(_a))
    p3 = (p1[0] + imgSize*np.cos(_a+np.pi/2.0)/2.0, p1[1] + imgSize*np.sin(_a+np.pi/2.0)/2.0)
    drawSineWare(axs[0], Isine, p3, angle)
       
        
    M = generateFourierMask(imgSize, angle)
    #IFrr = np.fft.fftshift( np.fft.fft2(I-np.mean(I)) )*M
    print(IFrr.shape, M.shape)
    IFrr = IFrr*M
    
    axs[1].imshow(np.log(1.0+np.abs(IFrr)))
        
    axs[1].plot([imgCenter[0] - 0.98*imgSize*np.cos(-angle)/2.0, imgCenter[0] + 0.98*imgSize*np.cos(-angle)/2.0],
             [imgCenter[1] - 0.98*imgSize*np.sin(-angle)/2.0, imgCenter[1] + 0.98*imgSize*np.sin(-angle)/2.0],'b')
    axs[1].plot([imgCenter[0] - w0*0.98*imgSize*np.cos(-angle)/2.0],
             [imgCenter[1] - w0*0.98*imgSize*np.sin(-angle)/2.0],'o',color=[1.0, 1.0, 1.0],markersize=10)
        

    axs[3].imshow(sino)
        
    
    Isine2 = generate2DSineWave(N, w0*imgSize/2.0, angle)
    IProd = I*Isine2
    axs[2].imshow(I*Isine2)
    
    
    fourierSlice = np.fft.fftshift( np.log(1.0+ np.abs(np.fft.fft(proj-np.mean(proj))) ))
    axs[3].plot([angle*sino.shape[1]/np.pi, angle*sino.shape[1]/np.pi],[0, sino.shape[0]],'r')
    axs[4].plot(fourierSlice)
    axs[4].plot([imgSize/2.0 + w0*imgSize/2.0],[fourierSlice[int(imgSize/2.0 + w0*(imgSize-1)/2.0)]],'o',color=[1.0, 0.0, 1.0],markersize=10)
    axs[4].plot([imgSize/2.0 + w0*imgSize/2.0, imgSize/2.0 + w0*imgSize/2.0],[0.0, fourierSlice[int(imgSize/2.0 + w0*(imgSize-1)/2.0)]],':',color=[1.0, 0.0, 1.0])
    
     
        
    axs[0].axis('equal') 
    axs[0].axis('off')
    axs[0].margins(0.0)
    axs[0].set_xlim([-3.0*imgSize+imgCenter[0], 3.0*imgSize+imgCenter[0]])
    axs[0].set_ylim([-3.0*imgSize+imgCenter[1], 3.0*imgSize+imgCenter[1]])
    
    axs[0].set_title('diagram')
    axs[1].set_title('2D Fourier')
    axs[2].set_title('I x sine')
    axs[3].set_title('projections')
    axs[4].set_title('Fourier Slice')
    
    
```

```python
def plotStateIteractive(I, IFtt, sino):
    fig = plt.figure(figsize=(20,10))
    fig.tight_layout()    

    axs = []
    axs.append( plt.subplot2grid(shape=(1, 2), loc=(0, 0)) )
    axs.append( plt.subplot2grid(shape=(2, 4), loc=(0, 3)) )
    axs.append( plt.subplot2grid(shape=(2, 4), loc=(0, 2)) )
    axs.append( plt.subplot2grid(shape=(2, 4), loc=(1, 2)) )
    axs.append( plt.subplot2grid(shape=(2, 4), loc=(1, 3)) )
    
    fig.subplots_adjust(left=0.05, bottom=0.05, right=0.95, top=0.95)

    def updatePlot(angleSliderValue = 0.0, w0SliderValue=0.0):
        [x.cla() for x in axs]
        plotState(axs, I, IFtt, sino, angleSliderValue, w0SliderValue)

    angleSlider = ipywidgets.widgets.FloatSlider(min=0.0, max=np.pi, step=0.1, value=0.0, layout=ipywidgets.Layout(width='100%'), continuous_update=True)
    w0Slider = ipywidgets.widgets.FloatSlider(min=-1.0, max=1.0, step=0.1, value=0.0, layout=ipywidgets.Layout(width='100%'), continuous_update=True)

    return ipywidgets.interact(updatePlot, angleSliderValue = angleSlider, w0SliderValue = w0Slider)
```

## Intuição e explicação gráfica

- A intuição pos tras do teorema pode ser ilustrada analisando-se alguns aspectos
- 1: Tome como exemplo uma projeção com um ângulo particular $\beta$
- 2: Use este mesmo ângulo para gerar as senoides 2D com frequencias $k_1$ e $k_2$ tais que o resultado seja uma senoide 2D inclinada exatamente de um ângulo $\beta$ e com uma frequencia na direção perpendicular a $\beta$ de valor igual a $\omega$
- 3: A transformada 2D de Fourier integra este produto para vários valores de $\omega$
- Ocorre que, para cada $\omega$, ao long de $\beta$ os valores da senoide 2D são constantes e seu efeito é apenas de "integrar a imagem original".
- Por esse motivo, em $\omega$, a transformada de Fourier não é mais a transformada de $f(x,y)$ e sim da integral de $f(x,y)$ ao longo de $\beta$. 
- Que é exatamente o que o teorema diz

## Extra: Filtered backprojection

### Porque backprojection funciona (ou funciona mais ou menos)

- Backprojection é o ato de tentar reconstruir a imagem operando com as projeções de maneira que, no final se projetarmos novamente, obtenhamos, ou $p_\beta(t)$ novamente ou alguma operação afim $c_1 + c_2\,p_\beta(t)$

- Uma forma de obter esta propriedade é fazer a média de funções formadas pelas projeções $p_i(x,y)$ onde $i=1,...,n$ são as $n$ projeções e $p$ é uma função que tem valor constante ao longo de $\beta_i$ e é igual a $p_{\beta_i}(t)$ na direção perpendicular a $\beta_i$. 

- Informalmente, $p_i(x,y)$ é uma "extrusão" de $p_\beta(t)$ ao longo e $\beta_i$ (ver figura)


<img src="backprojection.png"  width="800">


- Dessa maneira, claramente, a projeção de $p_i(x,y)$ ao longo de $\beta_i$ é exatamente a projeção original $p_{\beta_i}(t)$ (ver figura)

- Mas e as outras projeções? Para que essa estratégia funcionasse, a média de $p_j(x,y)$ ao longo de $\beta_i$ com $i\neq j$ deveria ser constante (para qualquer $\beta$ que não fosse o $\beta$ que gerou aquele $p_j(x,y)$...

- Em outras palavras, a transformada Radon de uma projeção "extrusa" é toda constante com exceção do ângulo desta projeção. Dessa maneira podemos somar as projeções que, no final, estaremos apenas somando uma constante à transformada Radon original

- Prova:

-- Podemos expressar $p_i(x,y)$ como uma função de apenas uma variável formando uma "extrusão" ao long de um ângulo $\beta$

$p_i(x,y) = g(x\,sin(\beta) + y\,cos(\beta))$

-- A média da sua projeção ao longo de um ângulo $\alpha$ é calculada como 

$ \lim_{a\to\infty} \frac{1}{2a} \int_{-a}^{a}{g( (t\,sin(\alpha) + \tau\,cos(\alpha))\,sin(\beta)  + (t\,cos(\alpha) - \tau\,sin(\alpha))\,cos(\beta)) \, \mathrm{d}\tau}$

-- Quando $\alpha = \beta$ temos:

$\lim_{a\to\infty} \frac{1}{2a} \int_{-a}^{a}{g( 
(t\,sin^2(\alpha) + \tau\,cos(\alpha)\,sin(\alpha))  +
(t\,cos^2(\alpha) - \tau\,sin(\alpha)\,cos(\alpha))
) \, \mathrm{d}\tau}$


$\lim_{a\to\infty} \frac{1}{2a} \int_{-a}^{a}{g( t ) \, \mathrm{d}\tau}$

-- que é igual a $g(t)$

-- Quando $\alpha \neq \beta$ podemos fazer a seguinte substituição de variáveis

$ \hat \tau =  (t\,sin(\alpha) + \tau\,cos(\alpha))\,sin(\beta)  + (t\,cos(\alpha) - \tau\,sin(\alpha))\,cos(\beta) $

$ \mathrm{d} \hat \tau =  sin(\beta-\alpha) \mathrm{d} \tau$


-- em que a integral fica


$ \lim_{a\to\infty} \frac{1}{2a} \int_{-a}^{a}{g( \hat \tau ) sin(\beta-\alpha) \mathrm{d} \tau}$

-- que é uma constnate independente de $t$


### Filtered backprojection

- Por causa do teorema de Fourier Slice, ao fazer a transformada Radon, estamos amostrando no dominio da frequencia em coordenadas polares

- Se fossemos usar a inversa de Fourier para recuperar a imagem original teríamos de faze-la em coordenadas polares. A transformada inverda de Fourier em coordenadas polares fica:

$
f(x,y) = \int_{-\pi}^{\pi}\int_{0}^{\infty}F(\omega\,cos(\beta),\omega\,sin(\beta))e^{j\omega\,(x\,cos(\beta) + y\,sin(\beta))}|\omega|\mathrm{d}\omega\mathrm{d}\beta
$

- Já que só amostramos $F(\omega\,cos(\beta),\omega\,sin(\beta))$ (por causa do teorema de slice de Fourier) precisamos multiplicar pelo termo $|\omega|$ antes de fazer a inversa.

- E esta é a razão pela qual precisamos de um filtro de "rampa", ou então derivar as projeções que é matematicamente igual a usar um filtro de rampa no dominio da freqüência.

- Na prática, não se faz a inversa nem se deriva as projeções. O que faz-se é convoluir com um filtro "derivativo" como laplace, sobel, etc.


## Fonte pontual (fanbeam)

- Nos casos reais de TC, não da para aplicar a transformada Radon diretamente porque os raios-X são emitidos de fontes pontuais

- Porem, para cada projeção da fonte pontual, existe uma projeção de uma fonte plana equivalente

- Dessa forma, basta escolhermos um mapeamento da projeção pontual para a projeção plana e utilizar a transformada Radon inversa.

- A figura a seguir ilustra um mapeamento possível. Nesta geometria, $t$ e $\beta$ são os parâmetros das projeções planas e $l$ e $\alpha$ os parâmetros das projeções pontuais.

<img src='radon_fanbeam.png' width=1000>


- Utilizando esse mapeamento, podemos, a partir de um sinograma gerado utilizando fontes pontuais (projeções com parâmetros $l$ e $\alpha$), mapear os valores de $t$ e $\beta$ gerando o sinograma equivalente à fonte plana.

<img src='radon_fanbeam_sample.png' width=1000>

- Note que a diferença entre a geometria dos pixels do detector podem gerar artefatos na imagem

- O mapeamento $M(S) = S(l, \alpha - \sin^{-1}(l/D))$ corresponde basicamente a uma operação de "shear" nos pixeis do sinograma

```python
def plotStateprojections(axs, beta, t):
    
    N = 128
    D = 2.0
    NProj = 9
    
    axs[0].plot([-D, D], [0.0, 0.0],'k')
    axs[0].plot([0.0, 0.0], [-D, D],'k')
    
    theta = np.linspace(-np.pi, np.pi, N)
    axs[0].plot(np.cos(theta), np.sin(theta))
    axs[0].plot([D*np.cos(beta) + np.sin(beta), 
                 D*np.cos(beta) - np.sin(beta)], 
                [D*np.sin(beta) - np.cos(beta), 
                 D*np.sin(beta) + np.cos(beta)],'g')
#     axs[0].plot([-D*np.cos(beta) + np.sin(beta), 
#                  -D*np.cos(beta) - np.sin(beta)], 
#                 [-D*np.sin(beta) - np.cos(beta), 
#                  -D*np.sin(beta) + np.cos(beta)],'g')

    for i in range(NProj):
        ti = -1.0 + 2.0*(i+0.5)/float(NProj)
        axs[0].plot([-D*np.cos(beta) + ti*np.sin(beta), 
                     D*np.cos(beta) + ti*np.sin(beta)], 
                    [-D*np.sin(beta) - ti*np.cos(beta), 
                     D*np.sin(beta) - ti*np.cos(beta)],'g', alpha=0.5)

        
    axs[0].plot([-D*np.cos(beta) + t*np.sin(beta), 
                  D*np.cos(beta) + t*np.sin(beta)], 
                [-D*np.sin(beta) - t*np.cos(beta), 
                  D*np.sin(beta) - t*np.cos(beta)],'m')
        
    axs[0].plot([0, D*np.cos(beta)], [0, D*np.sin(beta)],'r')
    
    axs[0].plot(0.3*np.cos(np.pi+theta[:int(beta/(2.0*np.pi)*N)]), 0.3*np.sin(np.pi+theta[:int(beta/(2.0*np.pi)*N)]),'k')

    axs[0].plot([0.0, t*np.sin(beta)], [0.0, -t*np.cos(beta)], 'm',alpha=0.5)
    
    axs[0].text(D*1.1*np.cos(beta), D*1.1*np.sin(beta),'source')
    axs[0].text(0.35*np.cos(beta), 0.35*np.sin(beta),'beta')
    axs[0].text(t*np.sin(beta)/2.0, -t*np.cos(beta)/2.0, 't')
    
    
    axs[0].axis('equal') 
    axs[0].axis('off')
    axs[0].margins(0.0)
    axs[0].set_xlim([-2.0, 2.0])
    axs[0].set_ylim([-2.0, 2.0])

    
    
    
    l = t
    alpha = beta - np.arcsin(l/D)
    
    
    axs[1].plot([-D, D], [0.0, 0.0],'k')
    axs[1].plot([0.0, 0.0], [-D, D],'k')
    
    theta = np.linspace(-np.pi, np.pi, N)
    axs[1].plot(np.cos(theta), np.sin(theta),'b')
    axs[1].plot([D*np.cos(alpha)], [D*np.sin(alpha)],'og')
#     axs[0].plot([-D*np.cos(beta) + np.sin(beta), 
#                  -D*np.cos(beta) - np.sin(beta)], 
#                 [-D*np.sin(beta) - np.cos(beta), 
#                  -D*np.sin(beta) + np.cos(beta)],'g')

    lMax = np.sqrt(D**2+1)
    for i in range(NProj):
        li = -lMax + 2.0*lMax*(i+0.5)/float(NProj)
        axs[1].plot([-D*np.cos(alpha) + li*np.sin(alpha), 
                     D*np.cos(alpha)], 
                    [-D*np.sin(alpha) - li*np.cos(alpha), 
                     D*np.sin(alpha)],'g', alpha=0.5)

        
    axs[1].plot([-D*np.cos(beta) + l*np.sin(beta), 
                  D*np.cos(beta) + l*np.sin(beta)], 
                [-D*np.sin(beta) - l*np.cos(beta), 
                  D*np.sin(beta) - l*np.cos(beta)],'m')
        
    axs[1].plot([0, D*np.cos(alpha)], [0, D*np.sin(alpha)],'r')
    
    axs[1].plot(0.3*np.cos(np.pi+theta[:int(alpha/(2.0*np.pi)*N)]), 0.3*np.sin(np.pi+theta[:int(alpha/(2.0*np.pi)*N)]),'k')

    axs[1].plot([0.0, l*np.sin(beta)], [0.0, -l*np.cos(beta)], 'm',alpha=0.5)
    
    axs[1].text(D*1.1*np.cos(alpha), D*1.1*np.sin(beta),'source')
    axs[1].text(0.35*np.cos(alpha), 0.35*np.sin(alpha),'alpha')
    axs[1].text(l*np.sin(alpha)/2.0, -l*np.cos(alpha)/2.0, 'l')
    
    
    axs[1].plot(np.cos(theta), np.sin(theta))
    
    
    
    
    
    
    axs[0].axis('equal') 
    axs[0].axis('off')
    axs[0].margins(0.0)
    axs[0].set_xlim([-1.5*D, 1.5*D])
    axs[0].set_ylim([-1.5*D, 1.5*D])    
    
    
    
    
    
    
    axs[1].axis('equal') 
    axs[1].axis('off')
    axs[1].margins(0.0)
    axs[1].set_xlim([-1.5*D, 1.5*D])
    axs[1].set_ylim([-1.5*D, 1.5*D])
                     
                     
    axs[0].set_title('planar source')
    axs[1].set_title('point source')
    
    
def plotProjections():
    fig = plt.figure(figsize=(10,6))
    fig.tight_layout()    

    axs = []
    axs.append( plt.subplot2grid(shape=(1, 2), loc=(0, 0)) )
    axs.append( plt.subplot2grid(shape=(1, 2), loc=(0, 1)) )
    
    fig.subplots_adjust(left=0.05, bottom=0.05, right=0.95, top=0.95)

    def updatePlot(betaSliderValue = 0.0, tSliderValue=0.0):
        [x.cla() for x in axs]
        plotStateprojections(axs, betaSliderValue, tSliderValue)

    betaSlider = ipywidgets.widgets.FloatSlider(min=0.0, max=2.0*np.pi, step=0.1, value=0.0, layout=ipywidgets.Layout(width='100%'), continuous_update=True)
    tSlider = ipywidgets.widgets.FloatSlider(min=-1.0, max=1.0, step=0.1, value=0.0, layout=ipywidgets.Layout(width='100%'), continuous_update=True)

    return ipywidgets.interact(updatePlot, betaSliderValue = betaSlider, tSliderValue = tSlider)    
```
