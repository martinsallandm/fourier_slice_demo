```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets
from functools import reduce
```

## Exemplo simples da transformada de Fourier 

Nesta primeira parte demonstramos como obter os coeficientes da transformada utilizando "integração" com senoides básicas

Algumas observações são importantes:

- Este exemplo é disreto, a transformada de Fourier possui diversas "versões" para sinais discretos e periódicos
- O nível de interpretação da transformada de Fourier pode tomar diversas formas, desde composição de sinais à representações em espaços vetoriais, produtos internos, etc

```python
N = 50

t = 2.0*np.pi*np.linspace(0, N-1, N)/N
x = t*np.exp(-(t-np.pi)**2)*np.sin(6*t) + np.sin(16*t)/(2*t+0.1) + 4*np.cos(t/4) - 2 

# X = np.fft.fft(x)

Xf = 1j*np.ones(N)
for w in range(N):
    s = np.sin(t*w)
    c = np.cos(t*w)
    
    e = c+1j*s
    
    Xf[w] = np.sum(x*e)
```

```python
plt.subplot(2,1,1)
plt.plot(t,x)
plt.title('sinal no tempo')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.subplot(2,1,2)
plt.plot(np.linspace(-np.pi,np.pi,Xf.shape[0]), np.fft.fftshift(np.abs(Xf)))
plt.title('sinal na frequencia')
plt.xlabel('w')
plt.ylabel('|X(w)|')
# plt.subplot(3,1,3)
# plt.plot(np.abs(X))
```

## Exemplo de transformada inversa de Fourier

Nesta parte demonstra-se a formação do sinal a partir das senoides individuais amplificadas e defasadas de acordo com os coeficiêntes calculados na transformada direta de Fourier

O sinal na vertical sào os coeficientes (módulo e fase) e os sinais na horizontal são as senoides individuais ja amplificadas e defasadas

```python
fig = plt.figure(figsize=(20,40))
fig.tight_layout()

colSpans = 3

axs = []
for i in range(N):
    axs.append( plt.subplot2grid(shape=(N+3, colSpans), loc=(i, 0),colspan = colSpans-1) )

ax = plt.subplot2grid(shape=(N+3, colSpans), loc=(0, colSpans-1),rowspan=N)
ax2 = plt.subplot2grid(shape=(N+3, colSpans), loc=(N, 0),colspan=colSpans-1, rowspan=3)

upsample = 1
xUp = np.ones(upsample*N)
tUp = 2.0*np.pi*np.linspace(0, upsample*N-1, upsample*N)/(upsample*N)
for i in range(N):
    s = np.sin(tUp*i)/N
    c = np.cos(tUp*i)/N

    xHatReal = c*np.real(Xf[i]) - s*np.imag(Xf[i])
    xHatImag = s*np.real(Xf[i]) + c*np.imag(Xf[i])
    
    xUp +=  xHatReal
    
    axs[i].plot(xHatReal+xHatImag)
    axs[i].plot([0,N-1],[0, 0],'k')
    axs[i].set_ylim([-1.4, 1.4])
    axs[i].axis('off')
    
XfAbs = reduce(lambda x,y:x+y, [[0,x,0] for x in np.abs(Xf[::-1])])
XfAngle = reduce(lambda x,y:x+y, [[0,x, 0] for x in np.angle(Xf[::-1])])
idx = reduce(lambda x,y:x+y, [[x,x,x] for x in range(N)])


ax.plot(XfAbs,idx,'o-')
ax.plot(XfAngle,idx,'o-')
ax.plot([0,0],[0, N],'k-')
ax.axis('off')
ax2.plot(xUp[::-1])
ax.set_ylim([-0.5,N-0.5])

```

```python
def reshape(I):
    if I.shape[0]>I.shape[1]:
        dif = I.shape[0]-I.shape[1]
        I = I[dif//2:-dif//2,:]    
    else:
        dif = I.shape[1]-I.shape[0]
        I = I[:dif//2:-dif//2]            
    
    return I
```

```python
def plotState(axs, I, IFrr, k1, k2):
    
    
    print(k1, k2)
    
    S = I.shape[0]
    
    clim = np.max(np.abs(IFrr))
    
    (X,Y) = np.meshgrid(np.linspace(-np.pi, np.pi, S), np.linspace(-np.pi, np.pi, S))
    
    sine2D = np.real(np.exp(1j*((k1-S/2.0)*X + (k2-S/2.0)*Y)))
    
    axs[0].imshow(I)
    axs[0].title('I')
    axs[1].imshow(np.log(1.0+np.abs(IFrr)))
    axs[1].plot([k1],[k2],'o')
    axs[1].title('F{I}')
    axs[2].imshow(sine2D*np.abs(IFrr[k1,k2]), clim=[-clim/10.0,clim/10.0])
    axs[2].title('sin(k1*x + k2*y)')
```

```python
def plotStateIteractive(I, IFtt):
    fig = plt.figure(figsize=(12,6))
    fig.tight_layout()    

    axs = []
    axs.append( plt.subplot2grid(shape=(1, 3), loc=(0, 0)) )
    axs.append( plt.subplot2grid(shape=(1, 3), loc=(0, 1)) )
    axs.append( plt.subplot2grid(shape=(1, 3), loc=(0, 2)) )
    
    fig.subplots_adjust(left=0.05, bottom=0.05, right=0.95, top=0.95)

    def updatePlot(k1SliderValue = 0, k2SliderValue=0):
        [x.cla() for x in axs]
        plotState(axs, I, IFtt, k1SliderValue, k2SliderValue)

    k1Slider = ipywidgets.widgets.IntSlider(min=0, max=I.shape[0], value=I.shape[0]//2, layout=ipywidgets.Layout(width='100%'), continuous_update=True)
    k2Slider = ipywidgets.widgets.IntSlider(min=0, max=I.shape[0], value=I.shape[0]//2, layout=ipywidgets.Layout(width='100%'), continuous_update=True)

    return ipywidgets.interact(updatePlot, k1SliderValue = k1Slider, k2SliderValue = k2Slider)
```

```python
%matplotlib qt5

I = reshape(plt.imread('euler_portrait.jpg')[:,:,0])
IFrr = np.fft.fftshift( np.fft.fft2(I-np.mean(I)) )
N = IFrr.shape[0]*IFrr.shape[1]

plotStateIteractive(I, IFrr)
```
